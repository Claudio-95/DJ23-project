```python
import pandas as pd

df = pd.read_csv("dataset_italianoEinglese_martina.csv")
df.head(5)
```

```python
df = df[df['lang'] == 'en']
df
```

## Preprocessing

```python
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import re

def clean_data(review):
    
    no_punc = re.sub(r'[^\w\s]', '', review)
    no_digits = ''.join([i for i in no_punc if not i.isdigit()])
    
    return(no_digits)

df['tweetText'] = df['tweetText'].apply(clean_data)
#df['Comment'] = df['Comment'].apply(clean_data)
```

```python
stop_words = set(stopwords.words('english'))  # Import english stop words
df['tweetText'] = df['tweetText'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))  # Remove stop words
```

## Sentiment analysis con NLTK

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
df['Sentiment'] = ''
for index, row in df.iterrows():
    comment = row['tweetText']
    
    # Applica VADER all'attuale commento
    scores = sia.polarity_scores(comment)
    compound_score = scores['compound']
    
    # Assegna un'etichetta di sentiment in base al punteggio compound
    if compound_score > 0:
        sentiment = 'pos'
    elif compound_score < 0:
        sentiment = 'neg'
    else:
        sentiment = 'neu'
    
    # Assegna l'etichetta di sentiment al DataFrame
    df.at[index, 'Sentiment'] = sentiment

# Stampa il DataFrame con le nuove etichette di sentiment
df['Sentiment'].value_counts()
```

## Sentiment analysis con TextBlob

```python
from textblob import TextBlob

# Crea una nuova colonna nel DataFrame per le etichette di sentiment
df['Sentiment'] = ''

# Itera attraverso ogni riga del DataFrame
for index, row in df.iterrows():
    comment = row['tweetText']
    
    # Applica l'analisi dei sentimenti di TextBlob all'attuale commento
    blob = TextBlob(comment)
    sentiment = blob.sentiment.polarity
    
    # Assegna un'etichetta di sentiment in base alla polarità
    if sentiment > 0:
        sentiment_label = 'pos'
    elif sentiment < 0:
        sentiment_label = 'neg'
    else:
        sentiment_label = 'neu'
    
    # Assegna l'etichetta di sentiment al DataFrame
    df.at[index, 'Sentiment'] = sentiment_label

# Stampa il DataFrame con le nuove etichette di sentiment
df['Sentiment'].value_counts()
```

```python
text = ' '.join(df['tweetText'])
text
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation

# Download delle stopwords se necessario
nltk.download('stopwords')

# Rimozione di punteggiatura e tokenizzazione
tokens = word_tokenize(text.lower())
tokens = [token for token in tokens if token not in stopwords.words('english') and token not in punctuation]
```

# Calcolo frequenza delle parole e dei n-grams

```python
from nltk.probability import FreqDist

# Calcolo della frequenza delle parole
freq_dist = FreqDist(tokens)
freq_dist
```

```python
sorted_freq = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
sorted_freq
```

```python
for word, freq in sorted_freq[:50]:
    print(f'{word}: {freq}')
```

```python
top_words = sorted_freq[:20]
top_words
```

```python
words = [word for word, freq in top_words]
frequencies = [freq for word, freq in top_words]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(words, frequencies)
plt.xticks(rotation=45)
plt.xlabel('Parole')
plt.ylabel('Frequenze')
plt.title('Prime 20 parole più frequenti')
plt.show()
```

```python
# Ordina le parole e le frequenze in ordine decrescente
sorted_freq = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)

# Estrai le prime 20 parole e le relative frequenze
top_words = [word for word, freq in sorted_freq[:20]]
top_frequencies = [freq for word, freq in sorted_freq[:20]]

# Crea il barplot
plt.figure(figsize=(8, 6))
plt.barh(top_words, top_frequencies)
plt.gca().invert_yaxis()  # Inverte l'asse y per posizionare la parola con la frequenza più alta in alto
plt.xlabel('Frequenze')
plt.ylabel('Parole')
plt.title('Prime 20 parole più frequenti')
plt.show()
```

# Trigrammi

```python
# Calcolo dei trigrammi
from nltk import ngrams
trigrams = list(ngrams(tokens, 3))
trigrams
```

```python
# Calcolo della frequenza dei trigrammi
freq_dist = FreqDist(trigrams)

# Trigrammi più frequenti
top_trigrams = freq_dist.most_common(50)

# Stampa dei trigrammi più frequenti
for trigram, frequency in top_trigrams:
    print(' '.join(trigram), ':', frequency)
```

```python
dfTrigram = pd.DataFrame(data=top_trigrams,columns=['words', 'frequency'])

dfTrigram
```

```python
import matplotlib.pyplot as plt

# Estrazione dei trigrammi e delle frequenze
trigrams = [' '.join(trigram) for trigram, frequency in top_trigrams]
frequencies = [frequency for trigram, frequency in top_trigrams]

# Configurazione del grafico a barre
plt.figure(figsize=(12, 8))
plt.barh(trigrams, frequencies)
plt.xlabel('Frequenza')
plt.ylabel('Trigrammi')
plt.title('Trigrammi più frequenti')
plt.gca().invert_yaxis()  # Inverte l'asse y per posizionare i trigrammi più frequenti in cima
plt.show()
```
