```python
import pandas as pd

df = pd.read_csv("dataset_italianoEinglese_martina.csv")
df.head(5)
```

```python
df = df[df['lang'] == 'it']
df
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace= True)
df
```

```python
#df.to_csv('chatGPT-dataset-martina.csv')
```

## Preprocessing

```python
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import re

def clean_data(review):
    
    no_punc = re.sub(r'[^\w\s]', '', review)
    no_digits = ''.join([i for i in no_punc if not i.isdigit()])
    
    return(no_digits)

df['tweetText'] = df['tweetText'].apply(clean_data)
#df['Comment'] = df['Comment'].apply(clean_data)
```

```python
stop_words = set(stopwords.words('italian'))  # Import english stop words
df['tweetText'] = df['tweetText'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))  # Remove stop words
```

## Sentiment analysis con NLTK

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
df['Sentiment'] = ''
for index, row in df.iterrows():
    comment = row['tweetText']
    
    # Applica VADER all'attuale commento
    scores = sia.polarity_scores(comment)
    compound_score = scores['compound']
    
    # Assegna un'etichetta di sentiment in base al punteggio compound
    if compound_score > 0:
        sentiment = 'pos'
    elif compound_score < 0:
        sentiment = 'neg'
    else:
        sentiment = 'neu'
    
    # Assegna l'etichetta di sentiment al DataFrame
    df.at[index, 'Sentiment'] = sentiment

# Stampa il DataFrame con le nuove etichette di sentiment
df['Sentiment'].value_counts()
```

```python
df.shape
```

## Sentiment distribution

```python
sizes = df["Sentiment"].value_counts()
labels = df["Sentiment"].unique()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [12, 8]

plt.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = False, startangle = 90)
plt.title("Sentiment distribution of italian Twitter dataset - NLTK")
plt.savefig("pie_nltksentiment-complete.png")
plt.show()
```

```python
from textblob import TextBlob

# Crea una nuova colonna nel DataFrame per le etichette di sentiment
df['Sentiment'] = ''

# Itera attraverso ogni riga del DataFrame
for index, row in df.iterrows():
    comment = row['tweetText']
    
    # Applica l'analisi dei sentimenti di TextBlob all'attuale commento
    blob = TextBlob(comment)
    sentiment = blob.sentiment.polarity
    
    # Assegna un'etichetta di sentiment in base alla polarità
    if sentiment > 0:
        sentiment_label = 'pos'
    elif sentiment < 0:
        sentiment_label = 'neg'
    else:
        sentiment_label = 'neu'
    
    # Assegna l'etichetta di sentiment al DataFrame
    df.at[index, 'Sentiment'] = sentiment_label

# Stampa il DataFrame con le nuove etichette di sentiment
df['Sentiment'].value_counts()
```
