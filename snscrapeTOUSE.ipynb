```python
import snscrape.modules.twitter as sntwitter
```

```python
# Setting variables to be used below
maxTweets = 1000000

loc = '41.29246 12.5736108, 200km'
# Creating list to append tweet data to
tweets_list2 = []
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('chatGPT geocode:"{}" since:2023-04-13 until:2023-05-13'.format(loc)).get_items()):
    if i>maxTweets:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.username, tweet.likeCount, tweet.lang, tweet.source, tweet.replyCount, tweet.mentionedUsers, tweet.outlinks])
```

```python
tweets_list2
```

```python
import pandas as pd
columns = ["Date Created", "Tweet ID", "Content", "Username", "Nr. Likes", "Lang", "Tweet Source", "Replies", "Mentions", "outlinks"]
#Creation of Dataframe
tweets_df = pd.DataFrame(tweets_list2, columns=columns)
```
